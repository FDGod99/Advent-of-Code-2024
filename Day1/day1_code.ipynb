```python
list1 = []
list2 = []

# Read the file
with open('/Users/fdgod/Desktop/FDGod/AdventOfCode2024/Day1/day1_input.txt', 'r') as file:
    for line in file:
        # Split the line into two numbers
        num1, num2 = map(float, line.split())  # Use `float` or `int` as per your data
        list1.append(num1)
        list2.append(num2)

print("List 1:", list1)
print("List 2:", list2)
```

```python
sorted_list1 = sorted(list1)
sorted_list2 = sorted(list2)
distance = 0
for i in range(len(sorted_list1)):
    distance = distance+abs(sorted_list1[i]-sorted_list2[i])
```

```python
print("Distance:", distance)
```

```python
similarity_score = 0
for i in range(len(sorted_list1)):
    count = 0
    for j in range(len(sorted_list2)):
        if sorted_list1[i] == sorted_list2[j]:
            count = count+1
    similarity_score = similarity_score+(count*sorted_list1[i])
```

```python
print("Similarity Score:", similarity_score)
```
